```python
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  
```

```python

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_Us = df[df['job_country'] == 'United States']
```

```python
df_skills = df_Us.explode('job_skills')
df_skills[['job_title', 'job_skills']]
```

```python
df_skills_count = df_skills.groupby(['job_skills', 'job_title_short']).size()
df_skills_count = df_skills_count.reset_index(name = 'Skills_count')
df_skills_count = df_skills_count.sort_values(by='Skills_count', ascending=False)
df_skills_count
```

```python
job_titles = df_skills_count['job_title_short'].unique().tolist()
job_titles = sorted(job_titles[:3])
job_titles
```

```python
fig, ax = plt.subplots(len(job_titles), 1)

for i, job_title in enumerate(job_titles):
    df_plot = df_skills_count[df_skills_count['job_title_short']== job_title].head(5)[::-1]
    sns.barplot(data=df_plot, x='Skills_count', y='job_skills', ax=ax[i], hue='Skills_count', palette = 'dark:b_r')
    ax[i].set_title(job_title)
    ax[i].invert_yaxis()
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].get_legend().remove()
    ax[i].set_xlim(0,45000)

fig.suptitle('Counts of skills required in US job Postings', fontsize =15)
plt.tight_layout(h_pad=0.5)
plt.show()
```

```python
df_job_title_count = df_Us['job_title_short'].value_counts().reset_index(name='job_total')
df_job_title_count
```

```python
df_skills_perc = pd.merge(df_skills_count, df_job_title_count, on='job_title_short', how='left')
df_skills_perc['skill_percent'] = (df_skills_perc['Skills_count']/df_skills_perc['job_total'])*100
df_skills_perc
```

```python
fig, ax = plt.subplots(len(job_titles), 1)

for i,job_title in enumerate(job_titles):
    df_plot = df_skills_perc[df_skills_perc['job_title_short'] == job_title].head(5)
    sns.barplot(data=df_plot, x='skill_percent', y='job_skills', ax=ax[i], hue='Skills_count', palette = 'dark:b_r')
    ax[i].set_title(job_title)
    ax[i].set_xlabel('')
    ax[i].set_ylabel('')
    ax[i].get_legend().remove()
    ax[i].set_xlim(0,78)

    if i != len(job_titles) -1: #len(job_titles) -1 represents the index of the last item
        ax[i].set_xticks([])

    for n, v in enumerate(df_plot['skill_percent']):
        ax[i].text(v+1, n,f'{v:.0f}%', va='center')

fig.suptitle('Likelihood of Skills Required in US Job Postings', fontsize = 15)
fig.tight_layout(h_pad=.8)
plt.show()    
```
